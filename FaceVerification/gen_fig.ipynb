```python
# discard some data by cutting off max and min mean of cosine features
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.svm import SVC

def load(i):
    with open(str(i)+'.pkl', 'rb') as f:
        d = pickle.load(f)
    return d['x'],d['y'],d['xt'],d['yt']

def get_cos(x, y, xt, yt):
    #print('   raw', x.shape, y.shape, xt.shape, yt.shape)
    xx, xxt = [], []
    for i in range(x.shape[0]):
        xx.append(x[i][x[i]<1])
    for i in range(xt.shape[0]):
        xxt.append(xt[i][xt[i]<1])
    return np.array(xx), y, np.array(xxt), yt

def SVM(x, y, xt, yt):
    clf = SVC(gamma='auto', probability=True)
    clf.fit(x,y)
    a = clf.score(x,y)
    b = clf.score(xt,yt)
    #print('train acc ', a)
    #print('test acc ',b)
    return a, b

def discard(x, y, xt, yt, percent=0.1):
    mx = np.mean(x, axis=1)
    idx = np.argsort(mx)
    s = (int)(idx.shape[0]*(percent))//2
    idx = idx[s:-s]
    return idx

per = []
ax = np.zeros((10, 30))
axt = np.zeros((10, 30))
for i in range(1,11):
    print(i)
    for j in range(1,31):
        per.append(1-j/100)
        x, y, xt, yt = load(i)
        x1, y1, xt1, yt1 = get_cos(x, y, xt, yt)
        idx = discard(x1, y1, xt1, yt1, 1-j/100)
        print('   train', x[idx].shape, y[idx].shape, xt.shape, yt.shape, i, j)
        #with open(str(i)+'_discard.pkl','wb') as f:
        #    pickle.dump({'x':x[idx], 'y':y[idx],
        #                    'xt':xt, 'yt':yt}, f, protocol=4)
        ax[i-1,j-1], axt[i-1,j-1] = SVM(x[idx], y[idx], xt, yt)
        per.append(x[idx].shape[0])
        #break
    #break


print('test acc')
print(axt)
```
